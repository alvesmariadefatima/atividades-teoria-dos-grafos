```python
def Top(n, Le, Ls):
  # inicia L como lista vazia
  L = []
  S = []
  #para cada um dos vértices
  for i in range(n):
    #FAZER (substituir null): se i não tem arestas de entrada:
    if (len(Le[i]) == 0):
      #adiciona o vértice i a lista S
      S.append(i)
  # enquanto S é não vazio
  while S:
    #remova um vértice de S
    m = S.pop()
    # FAZER: insira m em L
    L.append(m)

    # para cada vizinho de saida de m, remova a aresta
    while Ls[m]:
      v = Ls[m].pop()
      Le[v].remove(m)
      #FAZER (substituir null): se v não tem arestas de entrada:
      if (len(Le[v]) == 0):
        #FAZER: adicionar o vertice v a lista S
        S.append(v)
        
  #FAZER: ache uma maneira para verificar se ainda existe arestas no grafo
  # se ainda existir arestas coloque True na variavel aresta, coloque False caso contrário

  aresta = False
  for i in range(n):
    if len(Le[i]) > 0:
       aresta = True

  if aresta:
      print("Erro, o grafo possui ciclos, logo não tem ordenação topológica")
  else:
      print("Ordenação topológica proposta:")
      print(L)
```
